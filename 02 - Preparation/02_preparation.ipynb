### Boundary conditions and imports

```python
import pandas as pd
```

### Data Load

```python
# Create a sample DataFrame
data = pd.read_csv('dataset.csv')
data.head()
```

### General information

```python
# Review data information
data.info()
```

### Data manipulation

```python
# Found missing data and handle
data.isnull().sum()
data['goles_local'] = data['goles_local'].fillna(data['goles_local'].median())
print('Null values before injection: ')
print(data.isnull().sum())
```

```python
# One-Hot encoding
data_preparation = pd.get_dummies(data, columns=['equipo_local', 'equipo_visitante'])
data_preparation.head()
```

```python
# Delete duplicates
data_preparation.drop_duplicates(inplace=True)
print('Duplicates after deletion: ', len(data_preparation))
```

```python
# Manipulate dates
data_preparation['fecha_partido'] = pd.to_datetime(data_preparation['fecha_partido'], errors='coerce')
print('Data types after date manipulation: ', data_preparation['fecha_partido'].isnull().sum())
```
